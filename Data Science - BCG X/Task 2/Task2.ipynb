```python
sns.lineplot(data=price_df, x='price_date', y='price_off_peak_var', label='Off-Peak Var')
sns.lineplot(data=price_df, x='price_date', y='price_peak_var', label='Peak Var')
sns.lineplot(data=price_df, x='price_date', y='price_mid_peak_var', label='Mid-Peak Var')
sns.lineplot(data=price_df, x='price_date', y='price_off_peak_fix', label='Off-Peak Fix')
sns.lineplot(data=price_df, x='price_date', y='price_peak_fix', label='Peak Fix')
sns.lineplot(data=price_df, x='price_date', y='price_mid_peak_fix', label='Mid-Peak Fix')

plt.xlabel('Price Date')
plt.ylabel('Price')
plt.xticks(rotation = 60)
plt.title('Price Variations Over Time')
```
