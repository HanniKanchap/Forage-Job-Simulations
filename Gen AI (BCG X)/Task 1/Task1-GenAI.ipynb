```python
df = df.sort_values(by = ['Company','Fiscal Year']).reset_index()
```

```python
df.drop('index',axis = 1,inplace = True)
```

```python
df['Revenue Growth (%)'] =  df.groupby(['Company'])['Total Revenue'].pct_change()*100
```

```python
df['Net Income  Growth (%)'] =  df.groupby(['Company'])['Net Income '].pct_change()*100
```

# Revenue Growth and Net Income Growth Analysis

## Apple
- In **2023**, Apple's revenue declined by **2.79%**, and net income decreased by **2.91%**.
- In **2024**, revenue slightly recovered with a **2.01%** growth, but net income continued declining by **3.30%**.
- The continuous drop in net income suggests increasing costs or shrinking profit margins, despite revenue stabilization.

## Microsoft
- In **2023**, Microsoft saw a **6.86%** increase in revenue, but net income slightly dropped by **0.41%**.
- In **2024**, revenue surged by **15.67%**, accompanied by a strong net income growth of **21.69%**.
- This indicates Microsoft's ability to drive higher profitability alongside revenue expansion, potentially due to efficiency improvements or successful business strategies.

## Tesla
- In **2023**, Tesla experienced a **18.79%** revenue increase, alongside a **19.35%** rise in net income.
- However, in **2024**, revenue growth slowed dramatically to **0.95%**, while net income plunged by **52.70%**.
- The sharp decline in net income suggests major profitability challenges, possibly linked to operational costs or external market factors.
